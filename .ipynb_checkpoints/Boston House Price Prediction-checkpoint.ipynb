```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


%matplotlib inline
```

We can import the dataset directly from sklearn rather than downloading it to our harddrives.
Then load it to the notebook using pandas

```python
from sklearn.datasets import load_boston 
load_boston = load_boston() 
X = load_boston.data #This is the input data
y = load_boston.target #This is the output target

data = pd.DataFrame(X, columns=load_boston.feature_names) 
data["SalePrice"] = y # We created a new column called saleprice for the output target
data.head()
```

Let's describe the data before we get started
